```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
medical=pd.read_csv('/content/archive (3).zip')
medical.shape
```

```python
medical.head(5)
```

```python
sns.catplot(x ="sex", hue ="smoker", kind ="count", data = medical)
```

```python
sns.violinplot(x ="sex", y ="age", hue ="smoker",
data = medical, split = True)
```

```python
plt.scatter(medical["age"],medical["bmi"])
```

```python
sns.pairplot(medical)
```

```python
for col in medical.columns:
    sns.boxplot(
        data = medical,
        x = medical[col]
    )
    plt.show()
```

```python
group = medical.groupby(['sex', 'smoker'])
medical_ins = group.size().unstack()

# Heatmap - Color encoded 2D representation of data.
sns.heatmap(medical_ins, annot = True, fmt ="d")
```
